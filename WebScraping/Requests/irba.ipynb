```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
url = 'https://www.irba.co.za/find-an-ra?page=1&ipp=10&'
data = requests.get(url).text
```

```python
soup = BeautifulSoup(data, "html.parser")
tables = soup.find_all('table')
len(tables)
```

```python
#firstTable
auditors = tables[0]
body1 = auditors.find_all('tr')
head1 = body1[1]


headings1 = []

for item in head1.find_all('th'):
    item = (item.text).rstrip('\n')
    headings1.append(item)

body_rows1 = body1[2:-1]

all_rows1 = []
for row_num in range(len(body_rows1)):
    row = []
    for row_item in body_rows1[row_num].find_all('td'):
        row_item = re.sub("(\n)|,", "", row_item.text)
        row.append(row_item)
    all_rows1.append(row)

auditors_df = pd.DataFrame(data=all_rows1, columns=headings1)
auditors_df
```

```python
#secondTable
firms = tables[1]
body2 = firms.find_all('tr')
head2 = body2[1]


headings2 =[]
for item2 in head2.find_all('th'):
    item2 = (item2.text).rstrip('\n')
    headings2.append(item2)

body_rows2 = body2[2:-1]

all_rows2 = []
for row_num2 in range(len(body_rows2)):
    rows2 = []
    for row_item2 in body_rows2[row_num2].find_all('td'):
        row_item2 = re.sub("(\n|),", "", row_item2.text)
        rows2.append(row_item2)
    all_rows2.append(rows2)

firms_df= pd.DataFrame(data=all_rows2, columns=headings2)
firms_df
```
