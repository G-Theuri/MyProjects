```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
import tqdm.notebook as tqdm
import time
import pandas as pd
import json
```

```python
driver = webdriver.Chrome()
driver.get('https://www.seafoodsource.com/supplier-directory')
time.sleep(20)

driver.find_element(By.XPATH, "//button[@id='onetrust-accept-btn-handler']").click()
```

```python
data = []
nextbutton =True


while nextbutton:
    listings = driver.find_elements(By.XPATH, "//ol[@class='ais-Hits-list']/li")
    time.sleep(1)
    for listing in listings:
        driver.execute_script("arguments[0].scrollIntoView();", listing)
        time.sleep(2)

        link = listing.find_element(By.XPATH, ".//div[@class='results-btn-wrap']/a").get_attribute('href')
        page = driver.execute_script("window.open('{}', '_blank');".format(link))
        driver.switch_to.window(driver.window_handles[1])
        
        data.append({
            'webpage': (driver.find_element(By.XPATH,"//div[@class='t3p0-col col-sm-12']/div/a").get_attribute('href') if not None else None),
            'email': (driver.find_element(By.XPATH,"//div[@class='t3p0-col col-sm-12']/div[2]/a").get_attribute('href') if not None else None),
            'phone': (driver.find_element(By.XPATH,"//div[@class='t3p0-col col-sm-12']/div[3]").text if not None else None),
            'address': (driver.find_element(By.XPATH,"//div[@class='t3p0-row row supplier-directory__address']/div").text if not None else None),
            #'category': (driver.find_element(By.XPATH,"//div[@class='t3p0-col col-sm-8']/div[2]/ul/li").text if not None else None),
            'event': (driver.find_element(By.XPATH,"//div[@class='t3p0-col col-sm-8']/div[3]").text if not None else None),
            })
        


        time.sleep(1)
        driver.close()

        driver.switch_to.window(driver.window_handles[0])
        time.sleep(2)
    try:
        next_page = driver.find_element(By.XPATH,"//a[@aria-label='Next']").click()
    except:
        nextbutton=False

with open('seafood.json', 'w') as f:
    json.dump(data, f)

driver.quit()
```
